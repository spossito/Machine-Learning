# **Library**

```python
# Import

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras 
from tensorflow.keras.layers import Dense 

# Version

print('Keras Version: {}'.format(keras.__version__))
print('Tensorflow Version: {}'.format(tf.__version__))
```

# **Dataset**

```python
# Variables 

xTrain = np.array(object = [-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
yTrain = np.array(object = [-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)

# Visualize

print('Independent Variables Shape: {}'.format(xTrain.shape))
print('Dependent Variable Shape: {}'.format(yTrain.shape))
```

# **Model**

```python
# Model 

model = keras.Sequential([Dense(units = 1, activation = None, input_shape = [1], name = 'Dense_1')])

# Compile 

model.compile(optimizer = 'sgd', loss = 'mean_squared_error')

# Training 

history = model.fit(x = xTrain, y = yTrain, batch_size = 1, epochs = 500)
```

# **Training Resume**

```python
# Visualize 

plt.figure(figsize = (8, 8))

graph = history.history

plt.plot(history.epoch, graph['loss'], c = 'red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss vs Epoch')
plt.show()
```

# **Inference**

```python
# Result Inference

value = [10.0]

prediction = model.predict([value], verbose = False)

# Visualize 

print('Inference Value is: {:.2f}'.format(prediction[0][0]))
```
